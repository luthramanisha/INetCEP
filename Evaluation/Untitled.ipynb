```python
import xlsxwriter
from datetime import datetime
workbook = xlsxwriter.Workbook('QueryOutput.xlsx')
worksheet = workbook.add_worksheet()
expenses = []
with open("Q8") as f:
    for line in f:
        if line.strip():
            splitLine=line.split(",")
            newLine = []
            for i in range(0,len(splitLine)):
                print(splitLine[i])
                #if i != 1:
                if i != 1:
                    newLine.append(splitLine[i].replace('\n', ' ').replace('\r', ''))
                else:
                    newLine.append(splitLine[i].split()[1])
            expenses.append(tuple(newLine))
            
# Some data we want to write to the worksheet.
#expenses = (
#    ['Rent', 1000],
#    ['Gas',   100],
#    ['Food',  300],
#    ['Gym',    50],
#)

# Start from the first cell. Rows and columns are zero indexed.
row = 0
col = 0

format5 = workbook.add_format({'num_format': 'hh:mm:ss.SS'})

# Iterate over the data and write it out row by row.
for runID, time, endToEndDelay, opGCreation, nodeDiscovery, plAD, selPath, overallCost, energy, bdp in (expenses):
    worksheet.write(row, col,     int(runID))
    worksheet.write(row, col + 1, datetime.strptime(time, '%H:%M:%S.%f'),format5)
    worksheet.write(row, col + 2, int(endToEndDelay))
    worksheet.write(row, col + 3, int(opGCreation))
    worksheet.write(row, col + 4, int(nodeDiscovery))
    worksheet.write(row, col + 5, int(plAD))
    worksheet.write(row, col + 6, selPath)
    worksheet.write(row, col + 7, float(overallCost))
    worksheet.write(row, col + 8, float(energy))
    worksheet.write(row, col + 9, float(bdp))
    row += 1


workbook.close()
```
